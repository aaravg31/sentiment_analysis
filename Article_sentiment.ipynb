## Import Libraries and Initialize API Client

```python
from newsapi import NewsApiClient
import requests
from bs4 import BeautifulSoup
import pandas as pd
import seaborn as sns
from nltk.sentiment import SentimentIntensityAnalyzer
import os
import hashlib
from config import NEWSAPI_KEY
import matplotlib.pyplot as plt

newsapi = NewsApiClient(api_key=NEWSAPI_KEY)
```

```python
# Define the company and search query
#Company_names : Infosys, TCS, SBI, L&T
company = 'L&T' 
search_query = company + ' shares'

# Fetch articles
articles = newsapi.get_everything(q=search_query, language='en', sort_by='relevancy')

filename = "stored_sentiments.csv"
```

## Define Helper Functions

Define several helper functions to process the articles, perform sentiment analysis, and store the results.

```python
## Function to create a hash of the text content
def hash_text(text, company):
    return hashlib.md5((company + text).encode('utf-8')).hexdigest()

# Function to perform sentiment analysis using VADER
def sentiment_analysis(df):
    sia = SentimentIntensityAnalyzer()
    df['Sentiment'] = df['Text'].apply(lambda text: sia.polarity_scores(text)['compound'])
    return df

# Function to fetch text content from a URL
def get_text_from_url(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        paragraphs = soup.find_all('p')
        return [para.get_text() for para in paragraphs]
    else:
        print(f"Failed to retrieve content from {url}")
        return ""

# Function to create a DataFrame from the filtered articles
def make_table(filtered_article_info):
    unique_hashes = set()

    data = []
    for company, url, date, text, text_hash in filtered_article_info:
        if text_hash not in unique_hashes:
            data.append({'Company': company, 'Date': date, 'URL': url, 'Text': text, 'Text Hash': text_hash})
            unique_hashes.add(text_hash)
    
    df = pd.DataFrame(data, columns=['Company','Date', 'URL', 'Text', 'Text Hash'])
    df['Date'] = pd.to_datetime(df['Date'])
    df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
    df = sentiment_analysis(df)
    return df  
    
# Function to store data to a CSV file
def store_data(new_df, filename):
    if os.path.exists(filename):
        existing_df = pd.read_csv(filename)
        if existing_df.empty:
            combined_df = new_df
        else:
            combined_df = pd.concat([existing_df, new_df], ignore_index=True)
    else:
        combined_df = new_df
    
    combined_df.to_csv(filename, index=False)

# Function to filter out articles that have already been processed
def filter_existing_articles(article_info, filename, company):
    if os.path.exists(filename):
        existing_df = pd.read_csv(filename)
        existing_hashes = set(existing_df['Text Hash'])
    else:
        existing_hashes = set()
    
    filtered_article_info = []
    for url, date in article_info:
        paragraphs = get_text_from_url(url)
        specific_paragraphs = [para for para in paragraphs if company.lower() in para.lower()]
        if specific_paragraphs:
            text = '\n'.join(specific_paragraphs)
            text_hash = hash_text(text,company)
            if text_hash not in existing_hashes:
                filtered_article_info.append((company, url, date, text, text_hash))
                existing_hashes.add(text_hash)
    
    return filtered_article_info
```

## Fetch, Filter, and Store Articles

```python
article_info = [(article['url'], article['publishedAt']) for article in articles['articles']]

filtered_article_info = filter_existing_articles(article_info, filename, company)

df = make_table(filtered_article_info)

store_data(df,filename)
```

## Plot Results

```python
# Function to plot sentiment over time for each company
def plot_sentiments_over_time(df):
    unique_companies = df['Company'].unique()
    num_companies = len(unique_companies)
    num_rows = (num_companies + 1) // 2 

    df = df.sort_values(by='Date')
    
    plt.figure(figsize=(24, 6 * num_rows))
    for i, company in enumerate(unique_companies, 1):
        plt.subplot(num_rows, 2, i) 
        sns.lineplot(data=df[df['Company'] == company], x='Date', y='Sentiment', marker='o')
        plt.title(f'Sentiment Over Time for {company}')
        plt.xlabel('')
        plt.ylabel('Sentiment Score')
        plt.xticks(rotation=45)
        plt.grid(True)
    plt.tight_layout()
    plt.show()

# Function to plot sentiment distribution for each company
def plot_sentiment_distribution(df):
    unique_companies = df['Company'].unique()
    num_companies = len(unique_companies)
    num_rows = (num_companies + 1) // 2 
    
    plt.figure(figsize=(24, 6 * num_rows))
    for i, company in enumerate(unique_companies, 1):
        plt.subplot(num_rows, 2, i) 
        sns.histplot(df[df['Company'] == company]['Sentiment'], bins=20, kde=True, color='blue')
        plt.title(f'Sentiment Score Distribution for {company}')
        plt.xlabel('Sentiment Score')
        plt.ylabel('Frequency')
        plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv(filename)
plot_sentiments_over_time(df)
```

```python
plot_sentiment_distribution(df)
```
